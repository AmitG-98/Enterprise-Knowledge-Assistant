```python
from langchain_community.document_loaders import DirectoryLoader, TextLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_community.vectorstores import FAISS

# Load
loader = DirectoryLoader("docs/", glob="**/*.txt", loader_cls=TextLoader)
raw_docs = loader.load()
print(f"Loaded: {len(raw_docs)} docs")

# Chunk
splitter = RecursiveCharacterTextSplitter(
    chunk_size=300,
    chunk_overlap=50,
    separators=["\n\n", "\n", ".", " "]
)
chunks = splitter.split_documents(raw_docs)
print(f"Chunks: {len(chunks)}")

# Index
vectorstore = FAISS.from_documents(chunks, embeddings)
vectorstore.save_local("faiss_index")
print(f"Vectors indexed: {vectorstore.index.ntotal}")
```

```python
from langchain_community.retrievers.bm25 import BM25Retriever
from langchain_core.runnables import RunnableLambda

bm25 = BM25Retriever.from_documents(chunks)
bm25.k = 3
faiss_ret = vectorstore.as_retriever(search_kwargs={"k": 3})

def hybrid_retrieve(query):
    faiss_docs = faiss_ret.invoke(query)
    bm25_docs  = bm25.invoke(query)
    seen, merged = set(), []
    for doc in faiss_docs + bm25_docs:
        if doc.page_content not in seen:
            seen.add(doc.page_content)
            merged.append(doc)
    return merged[:4]

hybrid_retriever = RunnableLambda(hybrid_retrieve)
print("Hybrid retriever ready!")
```

```python
from langchain_core.prompts import PromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

prompt = PromptTemplate(
    template="""You are a helpful enterprise assistant.
Answer the question using ONLY the context below.
Be short, direct, and factual. Do not make up anything.

Context:
{context}

Question: {question}

Answer:""",
    input_variables=["context", "question"]
)

def format_docs(docs):
    return "\n\n".join(doc.page_content for doc in docs[:3])

rag_chain = (
    {
        "context": hybrid_retriever | format_docs,
        "question": RunnablePassthrough()
    }
    | prompt
    | llm
    | StrOutputParser()
)

print("RAG chain ready!")
```

```python
import gradio as gr

def ask(question, history):
    if not question.strip():
        return "", history

    docs   = hybrid_retriever.invoke(question)
    source = docs[0].metadata['source'].split("/")[-1] if docs else "unknown"
    answer = rag_chain.invoke(question)
    reply  = f"{answer}\n\n📄 Source: {source}"

    history.append({"role": "user",      "content": question})
    history.append({"role": "assistant", "content": reply})
    return "", history

with gr.Blocks(title="Enterprise Knowledge Assistant") as demo:

    gr.Markdown("# Enterprise Knowledge Assistant")
    gr.Markdown("Ask questions about HR policies, IT SOPs, compliance, and engineering guidelines.")

    chatbot = gr.Chatbot(
        height=450,
        type="messages",
        allow_tags=False,
        # removed show_copy_button — not supported in this Gradio version
    )

    with gr.Row():
        msg = gr.Textbox(
            placeholder="e.g. How many leave days do I get?",
            label="",
            scale=5
        )
        btn = gr.Button("Ask", variant="primary", scale=1)

    gr.Examples(
        examples=[
            "How many annual leave days do employees get?",
            "What is the P1 incident response time?",
            "When can I deploy to production?",
            "How long is customer data retained?",
            "What is the WFH policy?",
            "Who do I escalate to after L2?",
        ],
        inputs=msg
    )

    gr.Button("Clear", variant="secondary").click(
        fn=lambda: [],
        inputs=None,
        outputs=chatbot
    )

    msg.submit(ask, [msg, chatbot], [msg, chatbot])
    btn.click(ask,  [msg, chatbot], [msg, chatbot])

demo.launch(share=True, debug=False)
```
